```python
import json
from random import randint
```

```python
path = "/home/sivakrishna/Documents/jupyter/finetuning-test/interviewer_tuning/data/domain-data.txt"
file = open(path, 'r', encoding='utf-8')
content = file.read()
file.close()
```

```python
def clean_text(txt):
    ret_txt = ""
    ret_lst = []
    for i in range(len(txt)-1):
        if txt[i-2: i] == "-\n":
            ret_txt = ret_txt[:-1]
        if txt[i] == "-":
            if txt[i-1] != " ":
                ret_txt += ""
        elif txt[i] == "\n":
            if txt[i+1] == "\n":
                ret_txt += "\n"
            else:
                ret_txt += " "
        elif txt[i] == "•":
            ret_txt += "\n•"
        else:
            ret_txt += txt[i]
    return ret_txt
```

```python
clean_content = clean_text(content)
```

```python
def stitch_sent(sent_lst, st_idx, nd_idx):
    ret_txt = ""
    idx = st_idx
    while idx < nd_idx:
        ret_txt += sent_lst[idx]
        idx += 1
        
    return ret_txt

def text_to_dict(text):
    out_lst, out_dict = [], {}
    sent_lst = text.split(".")
    ln = len(sent_lst)
    
    for i in range(ln):
        try:
            if sent_lst[i][0] == " ":
                sent_lst[i] = sent_lst[i][1:]
        except:
            continue
            
    for i in range(1*ln):
        temp_dict = {}
        st_idx = randint(0, ln-10)
        nd_idx = randint(1, 5)
        
        temp_dict["prompt"] = stitch_sent(sent_lst, st_idx, st_idx+nd_idx)
        temp_dict["completion"] = stitch_sent(sent_lst, st_idx+nd_idx, st_idx+nd_idx+randint(2, 5))   

        out_lst.append(temp_dict)
    
    out_dict["data"] = out_lst
    return out_dict
```

```python
data_dict = text_to_dict(clean_content)
```
